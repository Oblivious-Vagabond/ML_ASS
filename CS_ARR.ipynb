```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
df_train = pd.read_csv('../train_data.nosync/train_data.csv', header=None)
df_val = pd.read_csv('./val_data.csv', header=None)
df_train.head()
```

```python
df_train.shape
```

```python
# Split the training data into feature and target
X_train = df_train.iloc[:,:-1].values
y_train = df_train.iloc[:,-1:].values.ravel()
X_val = df_val.iloc[:,:-1].values
y_val = df_val.iloc[:,-1:].values.ravel()
```

```python
for beat in X_train[:4]:
    plt.plot(beat)
plt.legend(y_train[:4])
plt.show()
```

```python
clf_lr = LogisticRegression()
clf_lr.fit(X_train, y_train)
y_prediction = clf_lr.predict(X_val)
print("Logistic Regressaion Classifier Accuracy: " + str(accuracy_score(y_val, y_prediction)))
```

```python
#Confusion matrix
c_matrix = confusion_matrix(y_val, y_prediction)
ConfusionMatrixDisplay(c_matrix).plot()
```

```python
clf_kn = KNeighborsClassifier(3)
clf_kn.fit(X_train, y_train)
y_prediction = clf_kn.predict(X_val)
print("K Neighbours Classifier Accuracy: " + str(accuracy_score(y_val, y_prediction)))
```

```python
#Confusion matrix
c_matrix = confusion_matrix(y_val, y_prediction)
ConfusionMatrixDisplay(c_matrix).plot()
```

```python
clf_mlp = MLPClassifier(hidden_layer_sizes=(100))
clf_mlp.fit(X_train, y_train)
y_prediction = clf_mlp.predict(X_val)
print("MLP Classifier Accuracy: " + str(accuracy_score(y_val, y_prediction)))
```

```python
#Confusion matrix
c_matrix = confusion_matrix(y_val, y_prediction)
ConfusionMatrixDisplay(c_matrix).plot()
```

```python
clf_mlp = MLPClassifier(hidden_layer_sizes=(50,50,50))
clf_mlp.fit(X_train, y_train)
y_prediction = clf_mlp.predict(X_val)
print("MLP Classifier Accuracy: " + str(accuracy_score(y_val, y_prediction)))
```

```python
#Confusion matrix
c_matrix = confusion_matrix(y_val, y_prediction)
ConfusionMatrixDisplay(c_matrix).plot()
```

```python
clf_mlp = MLPClassifier(hidden_layer_sizes=(100,100,100,100))
clf_mlp.fit(X_train, y_train)
y_prediction = clf_mlp.predict(X_val)
print("MLP Classifier Accuracy: " + str(accuracy_score(y_val, y_prediction)))
```

```python
#Confusion matrix
c_matrix = confusion_matrix(y_val, y_prediction)
ConfusionMatrixDisplay(c_matrix).plot()
```
